```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="/Users/Chorch/Desktop/Work/IronHack/Clases/SQL/Ironhack Data Analytics.json"
from google.cloud import bigquery
```

```python
royalty_sales = bigquery.Client()
QUERY ='''
SELECT
    tit_auth.title_id,
    tit_auth.au_id,
    (title_n.price * sales_n.qty * title_n.royalty / 100 * tit_auth.royaltyper / 100) AS price
FROM 
    `ironhack-data-analytics.publications.authors` auth
    
INNER JOIN 
    `ironhack-data-analytics.publications.titleauthor` tit_auth
ON auth.au_id = tit_auth.au_id

INNER JOIN 
    `ironhack-data-analytics.publications.titles` title_n
ON tit_auth.title_id = title_n.title_id

INNER JOIN 
    `ironhack-data-analytics.publications.sales` sales_n
ON sales_n.title_id = title_n.title_id

LIMIT 23
'''
```

```python
query_job = royalty_sales.query(QUERY)
df=query_job.to_dataframe()
df.head(10)
```
